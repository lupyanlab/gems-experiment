```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

landscape = pandas.read_csv('../gems/landscapes/SimpleHill.csv')
landscape.head()
```

```python
grid = landscape[['x', 'y', 'score']].pivot('x', 'y', 'score')
grid.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.contour3D(range(100), range(100), grid, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('score')
```
